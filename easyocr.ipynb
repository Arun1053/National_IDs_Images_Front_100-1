```python
import matplotlib.pyplot as plt
import cv2
import easyocr
import os
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize']=8,16
```

```python
reader = easyocr.Reader(['ar'])
```

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = os.path.join(folder,filename)
        if img is not None:
            images.append(img)
    return images
folder="/home/webwerks/Downloads/National_IDs_Images_Front_100-1/Images_100"
```

```python
load_images_from_folder(folder)
```

```python
abc = []
for i in load_images_from_folder(folder):
    output = reader.readtext(i)
    line = []
    for j in range(len(output)):
        line.append(output[j][1])
    abc.append(line)
```

```python
with open ('file1.txt','w',encoding='utf-8')as myfile:
	myfile.write(str(abc))
```

```python
import pandas as pd
data = pd.DataFrame(abc)
```
